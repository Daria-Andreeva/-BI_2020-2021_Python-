```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
df = pd.DataFrame(dict(Time=np.arange(500),
                       Value=np.random.randn(500).cumsum()))
g = sns.relplot(x="Time", y="Value", kind="line", data=df)
g.fig.autofmt_xdate()
```
